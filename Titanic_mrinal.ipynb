```python
import numpy as np
import pandas as pd
from google.colab import files

data = files.upload()
```

```python
df = pd.read_csv("train.csv")
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df['Age']=df['Age'].fillna(method='ffill')
```

```python
df.drop('Cabin', axis=1, inplace=True)
```

```python
df.info()
```

```python
common_value = 'S'
df["Embarked"] = df["Embarked"].fillna(common_value)
df["Embarked"].isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()
```

```python
df['Embarked']=le.fit_transform(df['Embarked'])
df.head()
```

```python
df.drop(['Name','Ticket'],axis=1, inplace=True)
```

```python
df['Sex']=le.fit_transform(df['Sex'])
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaled_fare = df['Fare']
scaled_fare = scaled_fare.values.reshape(-1, 1)
scaled_column = scaler.fit_transform(scaled_fare)
df['Fare'] = scaled_column
```

```python
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaled_age = df['Age']
scaled_age = scaled_age.values.reshape(-1, 1)
scaled_column = scaler.fit_transform(scaled_age)
df['Age'] = scaled_column
df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder as OHE
from sklearn.compose import ColumnTransformer
```

```python
ct = ColumnTransformer([('one_hot_encoder', OHE(categories='auto'), [8])],  remainder='passthrough')
x=ct.fit_transform(df)
```

```python
x
```

```python
df['Embarked_S'] = x[:,0]
df['Embarked_C'] = x[:,1]
df.head()
```

```python
df.Pclass.unique()
```

```python
df.drop(['Embarked'], axis=1, inplace=True)
```

```python
df.drop('PassengerId', axis=1, inplace=True)
```

```python
X=df.drop(['Survived'], axis=1)
Y=df['Survived']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_val, y_train, y_val = train_test_split(X, Y, test_size=0.2, random_state=1, stratify=Y)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

-------------------------------------------- LOGISTIC REGRESSION ----------------------------------

```python
lr = LogisticRegression(max_iter=1000).fit(x_train, y_train)

y_pred_lr = lr.predict(x_val)
print(lr.score(x_train, y_train))
print(accuracy_score(y_val,y_pred_lr))
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=6)
knn.fit(x_train, y_train)
y_pred = knn.predict(x_val)
score_knn = knn.score(x_val, y_val)
print(score_knn)
print(accuracy_score(y_val,y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier

dtc = DecisionTreeClassifier(criterion = 'log_loss', splitter = 'best', max_depth=16)
dtc = dtc.fit(x_train,y_train)
y_pred_dtc = dtc.predict(x_val)
score_dtc = dtc.score(x_val, y_val)
print(score_dtc)
print(accuracy_score(y_val,y_pred_dtc))
```

```python
from sklearn.svm import SVC
classifier = SVC(kernel = 'rbf', random_state = 0)
classifier.fit(x_train, y_train)
y_pred_classifier = classifier.predict(x_val)
from sklearn.metrics import accuracy_score
print(accuracy_score(y_val,y_pred_classifier))
print(classifier.score(x_train, y_train))
```

```python
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier
from sklearn.model_selection import StratifiedKFold, GridSearchCV

cv = StratifiedKFold(n_splits=5, random_state=0,shuffle=True)
extra=ExtraTreesClassifier()
param_grid = {
    "random_state": [0],
    "n_estimators": [360],
    "max_depth": [32]
}
searchCV = GridSearchCV(estimator=ExtraTreesClassifier(), scoring='accuracy', cv=cv, param_grid=param_grid, verbose=True)
searchCV.fit(X, Y)


print('Best index:', searchCV.best_index_)
print('Best score:', searchCV.best_score_)
print('Best params:', searchCV.best_params_)
```

```python
Extraclf = ExtraTreesClassifier(n_estimators=360, max_depth= 32, random_state=0)

Extraclf.fit(x_train, y_train)
y_pred_Extraclf= Extraclf.predict(x_val)
print(accuracy_score(y_val,y_pred_Extraclf))
print(Extraclf.score(x_train, y_train))
```

```python
neigh = KNeighborsClassifier(n_neighbors=210,weights='distance', algorithm='kd_tree',leaf_size=30 )
neigh.fit(x_train, y_train)
y_pred_neigh = neigh.predict(x_val)
print(accuracy_score(y_val,y_pred_neigh))
print(neigh.score(x_train, y_train))
```
